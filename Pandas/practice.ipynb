```python
import numpy as np 
import pandas as pd 
```

**Level 1: Series**

```python
# Create a Pandas Series from a list.
data = [10, 20, 30, 40, 50]
s = pd.Series(data)
s
```

```python
# Create a Series with custom indexes. 
data = [10, 20, 30, 40, 50]
s = pd.Series((data), index = ['A', 'B', 'C', 'D', 'E'])
s 
```

```python
# Create a Series from a dictionary. 
data = {
    "Math": 90,
    "Science": 85,
    "English": 88
}
s = pd.Series(data)
s
```

```python
# Print the first and last element of a Series. 
s = pd.Series([10, 20, 30, 40, 50])

print("First Element:", s.iloc[0])
print("Last Element :", s.iloc[-1])
```

```python
# Find the dtype, size, shape, and index of a Series.  
s = pd.Series([10, 20, 30, 40, 50])

print("Data Type:", s.dtype)
print("Size     :", s.size)
print("Shape    :", s.shape)
print("Index    :", s.index)
```

```python
# Calculate the sum, mean, maximum, and minimum of a Series. 
s = pd.Series([10, 20, 30, 40, 50])

print("Sum     :", s.sum())
print("Mean    :", s.mean())
print("Maximum :", s.max())
print("Minimum :", s.min())
```

```python
# Sort a Series by values. 
s = pd.Series([45, 10, 80, 25, 60])

print(s.sort_values()) 
```

```python
# Sort a Series by index.  
s = pd.Series([85, 90, 78],
              index=['C', 'A', 'B'])
print(s.sort_index()) 
```

**Level 2: DataFrame Creation**

```python
# Create a DataFrame from a dictionary.  
data = {
    "Name": ["John", "Alice", "Bob"],
    "Age": [25, 30, 28],
    "City": ["New York", "London", "Paris"]
}
df = pd.DataFrame(data)
df
```

```python
# Create a DataFrame from a list of lists.  
data = [
    ["Ram", 21, 85],
    ["Shyam", 22, 90],
    ["Mohan", 20, 78]
]

df = pd.DataFrame(data, columns=["Name", "Age", "Marks"])
df
```

```python
# Create a DataFrame from a list of dictionaries.  
data = [
    {"Name": "Ram", "Age": 21, "Marks": 85},
    {"Name": "Shyam", "Age": 22, "Marks": 90},
    {"Name": "Mohan", "Age": 20, "Marks": 78}
]
df = pd.DataFrame(data)
df
```

```python
# Create a DataFrame with custom indexes. 
data = [
    ["Ram", 21, 85],
    ["Shyam", 22, 90],
    ["Mohan", 20, 78]
]
df = pd.DataFrame(data, columns=["Name", "Age", "Marks"], index=["A", "B", "C"])
df
```

```python
# Print the first 5 rows using head().  
data = {
    "Name": ["Ram", "Shyam", "Mohan", "Amit", "Ravi", "Karan"],
    "Marks": [85, 90, 78, 88, 75, 95]
} 
df = pd.DataFrame(data)
print(df.head()) 
```

```python
# Print the last 3 rows using tail().  
data = {
    "Name": ["Ram", "Shyam", "Mohan", "Amit", "Ravi", "Karan"],
    "Marks": [85, 90, 78, 88, 75, 95]
} 
df = pd.DataFrame(data)
print(df.tail(3))  
```

```python
# Display dataset information using info()  
data = {
    "Name": ["Ram", "Shyam", "Mohan"],
    "Age": [21, 22, 20],
    "Marks": [85, 90, 78]
}
df = pd.DataFrame(data)
df.info()
```

```python
# Display statistical summary using describe()  
data = {
    "Age": [21, 22, 20, 23, 24],
    "Marks": [85, 90, 78, 88, 95]
}

df = pd.DataFrame(data)
print(df.describe())
```

**Level 3: DataFrame Attributes**

```python
# Find the shape of a DataFrame.  
data = {
    "Name": ["Ram", "Shyam", "Mohan"],
    "Age": [21, 22, 20],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print("Shape:", df.shape)
```

```python
# Find the column names.  
data = {
    "Name": ["Ram", "Shyam", "Mohan"],
    "Age": [21, 22, 20],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print("column names:", df.columns)
```

```python
# Find the row indexes.  
data = {
    "Name": ["Ram", "Shyam", "Mohan"],
    "Age": [21, 22, 20],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print("row indexes:", df.index)
```

```python
# Find the data types of all columns.  

data = {
    "Name": ["Ram", "Shyam", "Mohan"],
    "Age": [21, 22, 20],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print("data types of all columns is:", df.dtypes)
```

```python
# Find the total number of elements.  
data = {
    "Name": ["Ram", "Shyam", "Mohan"],
    "Age": [21, 22, 20],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print("total number of elements is:", df.size)
```

```python
# Convert a DataFrame into a NumPy array using values.  
data = {
    "Name": ["Ram", "Shyam", "Mohan"],
    "Age": [21, 22, 20],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print("DataFrame into a NumPy array using values:", df.values)
```

**Level 4: Selecting Data**

```python
# Sample DataFrame 
data = {
    "Name": ["Ram", "Shyam", "Mohan", "Amit", "Ravi", "Karan"],
    "Age": [21, 28, 25, 30, 27, 24],
    "Salary": [40000, 55000, 48000, 65000, 52000, 45000],
    "Department": ["HR", "IT", "Finance", "IT", "Sales", "IT"]
}

df = pd.DataFrame(data)

print(df) 
```

```python
# Select a single column.  
print(df["Name"])
```

```python
# Select multiple columns.  
print(df[["Name", "Age"]])
```

```python
# Select the first row using iloc  
print(df.iloc[0]) 
```

```python
# Select the first row using loc  
print(df.loc[0]) 
```

```python
# Select rows from index 2 to 5  
print(df.iloc[2:6]) 
```

```python
# Select rows where Age > 25  
print(df[df["Age"] > 25]) 
```

```python
# Select rows where Salary > 50000 
print(df[df["Salary"] > 50000])
```

```python
# Select rows where Department == "IT"  
print(df[df["Department"] == "IT"])
```

**Level 5: Missing Data**

```python
# Sample DataFrame  
data = {
    "Name": ["Ram", "Shyam", "Mohan", "Amit", "Ravi"],
    "Age": [21, np.nan, 25, 30, np.nan],
    "Salary": [40000, 55000, np.nan, 65000, 52000],
    "City": ["Nagpur", np.nan, "Pune", "Mumbai", np.nan]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Detect missing values using isnull()  
print(df.isnull())
```

```python
# Count missing values in each column.  
print(df.notnull())
```

```python
# Remove rows containing missing values.  
print(df.dropna())
```

```python
# Fill missing values with 0 
print(df.fillna(0))
```
