```python
import numpy as np 
import pandas as pd 
```

**Level 1: Series**

```python
# Create a Pandas Series from a list.
data = [10, 20, 30, 40, 50]
s = pd.Series(data)
s
```

```python
# Create a Series with custom indexes. 
data = [10, 20, 30, 40, 50]
s = pd.Series((data), index = ['A', 'B', 'C', 'D', 'E'])
s 
```

```python
# Create a Series from a dictionary. 
data = {
    "Math": 90,
    "Science": 85,
    "English": 88
}
s = pd.Series(data)
s
```

```python
# Print the first and last element of a Series. 
s = pd.Series([10, 20, 30, 40, 50])

print("First Element:", s.iloc[0])
print("Last Element :", s.iloc[-1])
```

```python
# Find the dtype, size, shape, and index of a Series.  
s = pd.Series([10, 20, 30, 40, 50])

print("Data Type:", s.dtype)
print("Size     :", s.size)
print("Shape    :", s.shape)
print("Index    :", s.index)
```

```python
# Calculate the sum, mean, maximum, and minimum of a Series. 
s = pd.Series([10, 20, 30, 40, 50])

print("Sum     :", s.sum())
print("Mean    :", s.mean())
print("Maximum :", s.max())
print("Minimum :", s.min())
```

```python
# Sort a Series by values. 
s = pd.Series([45, 10, 80, 25, 60])

print(s.sort_values()) 
```

```python
# Sort a Series by index.  
s = pd.Series([85, 90, 78],
              index=['C', 'A', 'B'])
print(s.sort_index()) 
```

**Level 5: Missing Data**

```python
# Sample DataFrame  
data = {
    "Name": ["Ram", "Shyam", "Mohan", "Amit", "Ravi"],
    "Age": [21, np.nan, 25, 30, np.nan],
    "Salary": [40000, 55000, np.nan, 65000, 52000],
    "City": ["Nagpur", np.nan, "Pune", "Mumbai", np.nan]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Detect missing values using isnull()  
print(df.isnull())
```

```python
# Count missing values in each column.  
print(df.notnull())
```

```python
# Remove rows containing missing values.  
print(df.dropna())
```

```python
# Fill missing values with 0 
print(df.fillna(0))
```

```python
# Fill missing values using the column mean.  
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
print(df) 
```

```python
# Replace missing city names with "Unknown"  
df["City"] = df["City"].fillna("Unknown")
print(df) 
```

**Level 6: Sorting & Filtering**

```python
# Display employees from the HR department.  
print(df[df["Department"] == "HR"]) 
```

**Level 7: Operations**

```python
# Sample DataFrame  
data = {
    "Name": ["Ram", "Shyam", "Mohan", "Amit", "Ravi", "Karan"],
    "Gender": ["Male", "Male", "Male", "Male", "Female", "Female"],
    "Marks": [65, 88, 75, 92, 70, 81],
    "Salary": [40000, 75000, 55000, 65000, 52000, 85000],
    "Department": ["HR", "IT", "Finance", "IT", "Sales", "HR"]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Add 10 to every value in the Marks column  
df["Marks"] = df["Marks"] + 10
print(df) 
```

```python
# Create a new column Bonus as 10% of Salary  
df["Bonus"] = df["Salary"] * 0.10
print(df) 
```

```python
# Convert all names to uppercase.  
df["Name"] = df["Name"].str.upper()
print(df) 
```

```python
# Replace "Male" with "M" and "Female" with "F".  
df["Gender"] = df["Gender"].replace({
    "Male": "M",
    "Female": "F"
})
print(df) 
```

```python
# Find all unique departments. 
print(df["Department"].unique()) 
```

```python
# Count the number of employees in each department.  
print(df["Department"].value_counts()) 
```

**Level 8: GroupBy & Aggregation**

```python
# Sample DataFrame  
data = {
    "Name": ["Ram", "Shyam", "Mohan", "Amit", "Ravi", "Karan"],
    "Department": ["HR", "IT", "Finance", "IT", "Sales", "HR"],
    "Salary": [40000, 75000, 55000, 65000, 52000, 85000]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Find the average salary of each department.  
print(df.groupby("Department")["Salary"].mean())
```

```python
# Find the maximum salary in each department.  
print(df.groupby("Department")["Salary"].max()) 
```

```python
# Count employees in each department.  
print(df.groupby("Department")["Name"].count()) 
```

```python
# Find the sum of salaries for each department.  
print(df.groupby("Department")["Salary"].sum()) 
```

```python
# Use agg() to calculate sum, mean, max, and min for salaries.  
print(df.groupby("Department")["Salary"].agg(["sum", "mean", "max", "min"])) 
```

**Level 9: Merge, Join & Concatenation**

```python
# Merge two DataFrames using a common ID column.  
df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Ram", "Shyam", "Mohan"]
})

df2 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Salary": [40000, 50000, 60000]
})

result = pd.merge(df1, df2, on="ID")

print(result)
```

```python
# Join two DataFrames using their indexes.  
df1 = pd.DataFrame({
    "Name": ["Ram", "Shyam", "Mohan"]
}, index=[1, 2, 3])

df2 = pd.DataFrame({
    "Salary": [40000, 50000, 60000]
}, index=[1, 2, 3])

result = df1.join(df2)

print(result) 
```

```python
# Concatenate two DataFrames vertically. 
df1 = pd.DataFrame({
    "Name": ["Ram", "Shyam"],
    "Salary": [40000, 50000]
})

df2 = pd.DataFrame({
    "Name": ["Mohan", "Amit"],
    "Salary": [60000, 70000]
})

result = pd.concat([df1, df2], axis=0)

print(result) 
```

```python
# Concatenate two DataFrames horizontally.  
df1 = pd.DataFrame({
    "Name": ["Ram", "Shyam", "Mohan"]
})

df2 = pd.DataFrame({
    "Salary": [40000, 50000, 60000]
})

result = pd.concat([df1, df2], axis=1)

print(result) 
```

**Level 10: Pivot Table**

```python
# Sample DataFrame  
data = {
    "Name": ["Ram", "Shyam", "Mohan", "Amit", "Ravi", "Karan"],
    "Department": ["HR", "IT", "Finance", "IT", "Sales", "HR"],
    "Gender": ["Male", "Male", "Male", "Male", "Female", "Female"],
    "City": ["Nagpur", "Pune", "Mumbai", "Pune", "Nagpur", "Mumbai"],
    "Salary": [40000, 75000, 55000, 65000, 52000, 85000],
    "Sales": [200, 500, 300, 400, 250, 350]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Create a pivot table showing the average salary by department.  
pivot = df.pivot_table(
    values="Salary",
    index="Department",
    aggfunc="mean"
)
print(pivot) 
```

```python
# Create a pivot table showing total sales by city.  
pivot = df.pivot_table(
    values="Sales",
    index="City",
    aggfunc="sum"
)

print(pivot) 
```

```python
# Create a pivot table with department as rows and gender as columns.  
pivot = df.pivot_table(
    values="Salary",
    index="Department",
    columns="Gender",
    aggfunc="mean"
)

print(pivot) 
```

```python
# Create a pivot table using sum as the aggregation function.  
pivot = df.pivot_table(
    values="Salary",
    index="Department",
    aggfunc="sum"
)

print(pivot)
```
